````python
# Cell 1: SQL Queries for Dashboard
print("""# 📊 Looker Studio Dashboard Setup Guide

## Step 1: Open Looker Studio
Buka https://lookerstudio.google.com

## Step 2: Create New Report
Click **Create** → **Report**

## Step 3: Add BigQuery Data Source
1. Click **Add data**
2. Search "BigQuery"
3. Select your project: `telco-portfolio`
4. Select dataset: `telco_dataset`
5. Select table: `features_network_kqi`
6. Click **Connect**

## Step 4: SQL Queries untuk Dashboard

### Query 1: Churn Rate by Network Quality
```sql
SELECT 
  CASE 
    WHEN network_quality_score < 30 THEN 'Poor Network'
    WHEN network_quality_score < 60 THEN 'Fair Network'
    WHEN network_quality_score < 80 THEN 'Good Network'
    ELSE 'Excellent Network'
  END AS network_tier,
  COUNT(*) as total_customers,
  SUM(churn_label) as churned,
  ROUND(100.0 * SUM(churn_label) / COUNT(*), 2) as churn_rate,
  ROUND(AVG(avg_throughput_mbps), 1) as avg_throughput,
  ROUND(AVG(latency_ms), 1) as avg_latency
FROM `telco-portfolio.telco_dataset.features_network_kqi`
GROUP BY network_tier
ORDER BY churn_rate DESC;

### Query 2: Monthly Charges vs Network Quality
```sql
SELECT 
  ROUND(MonthlyCharges, -1) as charge_bracket,
  ROUND(AVG(network_quality_score), 1) as avg_network_quality,
  COUNT(*) as customer_count,
  ROUND(AVG(churn_label) * 100, 2) as churn_rate
FROM `telco-portfolio.telco_dataset.features_network_kqi`
GROUP BY charge_bracket
ORDER BY charge_bracket;

### Query 3: Churn by Contract & Internet Service
```sql
SELECT 
  Contract,
  InternetService,
  COUNT(*) as customers,
  ROUND(AVG(churn_label) * 100, 2) as churn_rate,
  ROUND(AVG(network_quality_score), 1) as avg_quality
FROM `telco-portfolio.telco_dataset.features_network_kqi`
GROUP BY Contract, InternetService
ORDER BY churn_rate DESC;

## Step 5: Create Charts

1. **Scorecard**: Total Customers, Overall Churn Rate, Avg Network Quality
2. **Bar Chart**: Churn Rate by Network Quality Tier
3. **Line Chart**: Monthly Charges vs Network Quality
4. **Heatmap**: Churn Rate by Contract x Internet Service
5. **Table**: Top 10 customers with highest churn probability

## Step 6: Add Filter Controls
1. Add **Contract Type** filter
2. Add **Internet Service** filter  
3. Add **Network Quality Range** slider

## Step 7: Share Dashboard
Click **Share** → **Get report link** → Set to "Anyone with link can view"
""")
````
